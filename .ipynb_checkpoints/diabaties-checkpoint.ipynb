```python
df_cleaned = pd.get_dummies(df_cleaned,columns = ['diet_type'],drop_first = True)
```

```python
df_cleaned["diet_type_moderate"] = df_cleaned["diet_type_moderate"].astype(int)
df_cleaned["diet_type_unhealthy"] = df_cleaned["diet_type_unhealthy"].astype(int)
```

```python
df_cleaned.head()
```

```python
df_cleaned["diabetes_status"] = df_cleaned["diabetes_status"].map({"Low":0, "High":1})
```

```python
df_cleaned.head()
```

```python
df_cleaned['bmi_category'] = pd.cut(
    df_cleaned['bmi'],
    bins=[0, 18.5, 24.9, 29.9, float('inf')],
    labels=['underweight', 'normal', 'overweight', 'obese']
)
```

```python
df_cleaned.head()
```

```python
df_cleaned = pd.get_dummies(df_cleaned,columns = ['bmi_category'],drop_first = True)
```

```python
df_cleaned.head()
```

```python
df_cleaned["bmi_category_normal"] = df_cleaned["bmi_category_normal"].astype(int)
df_cleaned["bmi_category_overweight"] = df_cleaned["bmi_category_overweight"].astype(int)
df_cleaned["bmi_category_obese"] = df_cleaned["bmi_category_obese"].astype(int)
```

```python
df_cleaned.head()
```

```python
df_cleaned.columns
```

```python
from sklearn.preprocessing import StandardScaler
cols = ['age','bmi','fasting_glucose','systolic_bp','physical_activity']
scaler = StandardScaler()

df_cleaned[cols] = scaler.fit_transform(df_cleaned[cols])
```

```python
df_cleaned.head()
```

```python
X = df_cleaned.drop(columns=["diabetes_status"]) 
y = df_cleaned["diabetes_status"]
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,        # 20% for testing
    random_state=42,      # reproducible results
    stratify=y            # keep same diabetes distribution
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize model
clf = RandomForestClassifier(
    n_estimators=100, 
    random_state=42, 
    class_weight='balanced'  # helps with class imbalance
)

# Train
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
